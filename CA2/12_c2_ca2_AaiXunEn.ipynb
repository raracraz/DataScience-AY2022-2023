```python
# This box is used for showing the number of hospital admissions in Singapore due to Covid in the last 2 months

# Importing libraries #
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Code starts here #
# Importing the data #
df = pd.read_csv('./csvData/new-covid-19-hospital-admissions.csv')

# Processing the data #
df['date'] = pd.to_datetime(df['date'])
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
df['new_hospital_admissions'] = df['new_hospital_admissions'].astype(int)

# Visualizing the data #
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)
plt.title('Number of daily hospital admissions in Singapore due to Covid in the last 2 months', fontsize=20)

# Plotting the data #
sns.set_style('ticks', rc={'axes.facecolor': '.9', 'axes.grid': False})
major_ticks = np.arange(0, 221, 20)
minor_ticks = np.arange(0, 210, 5)
plt.xticks(rotation=60, ha='right')
ax.set_yticks(major_ticks)
ax.set_yticks(minor_ticks, minor=True)

ax.grid(which='minor', alpha=0.5, linestyle='dashed')
ax.grid(which='major', alpha=1)

sns.despine()
sns.barplot(x='date', y='new_hospital_admissions', data=df)

plt.ylabel('Number of hospital admissions', fontsize=20)
plt.xlabel('Dates recorded', fontsize=20)

# set grid on
plt.grid(True, 'major', 'y', ls='--', lw=.5, c='k', alpha=.3)

# set a best fit line to help see the trend between dates
ind = np.arange(len(df))
z = np.polyfit(ind, df['new_hospital_admissions'], 1)
p = np.poly1d(z)
plt.plot(ind, p(ind), "r--")
plt.show()

# Code ends here

# # # # # # # # # # # # # # # # # # # # # # # # # #  Analysis of the data and graph of the data # # # # # # # # # # # # # # # # # # # # # # # # # #
# IMPORTANT EVENTS in the 2 months:                                                                                                               #
# 1. Singapore to lift most lockdown restrictions on June 19 starting from June 2, 2022                                                           #
# 2. Group size limits will go up to 10 persons and masks will no longer be mandatory in outdoor settings from March 29 onwards                   #
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
print('OBSERVATIONS about the dataset:')
# show the shape of the dataset
print('\nThis is the shape of the dataset: ' + str(df.shape))
# show the index of the dataset
print('This is the index of the dataset: ' + str(df.index))
# show the columns of the dataset
print('This is the columns of the dataset: ' + str(df.columns))
# show a summary of the dataset
print('This is a summary of the dataset: ')
print(str(df.info()))
print('\nINSIGHTS:')
print('\nThe number of hospital admissions in Singapore due to Covid in the last 2 months is: ', str(df['new_hospital_admissions'].sum()) + '.')
print('The best fit line suggests that there is currently a trend of hospital admissions increasing in Singapore in the last 2 months becausing of Covid.')
# find the average number of hospital admissions increased per day 
print('The average number of hospital admissions increased per day in the last 2 months is: ', str((df['new_hospital_admissions'].mean() / 61).round(2)) + '.')
print('A significant spike of hospital admissions can be observed from June 19, 2022 onwards, likely caused by the lifting of lockdown restrictions and the increase in group size limits with no masks mandatory in outdoor settings and the lifting of safe distance measures.')
print('In order to move forward with the impact of Covid-19 in Singapore, the government lifted most restrictions; however, this resulted in the rise in trend of hospital admissions in the last 2 months. Some researchers have debated that this return to normality is too fast, and that the covid situation will worsen again if the government does lifts the lockdown restrictions, which came true as seen in the graph above.')
print('I feel that I would reccommend the government to have delayed the lifting of restrictions to when the covid situation gets better and reaches a more stable state.')
```

```python
# Percentage of groups of people being hospitalised, Needing oxygen supplementation, Admitted into ICU and Deceased
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# Importing the data
try:
    df = pd.read_csv('./csvData/patients-needing-oxygen-supplementation-icu-care-or-died-by-age-groups.csv')
    print('Loaded dataset successfully')
except:
    print('File not found')
# Loading the data into a dataframe
df['date'] = pd.to_datetime(df['day_of_as_of_date'])
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
df['age_groups'] = df['age_groups'].astype(str)
df['clinical_status'] = df['clinical_status'].astype(str)
df['count_of_case'] = df['count_of_case'].astype(int)
# Getting the total number of people hospitalised
df = df.groupby(['clinical_status']).sum().reset_index()
# if the clinical status is 'Hospitalised', minus all other clinical statuses to find the actual number of people hospitalised
total_hospitalised = df['count_of_case'][df['clinical_status'] == 'Hospitalised'].sum()
total_not_hospitalised = df['count_of_case'][df['clinical_status'] != 'Hospitalised'].sum()
actual_hospitalised = total_hospitalised - total_not_hospitalised
# Total people admitted to ICU
total_icu = df['count_of_case'][df['clinical_status'] == 'ICU'].sum()
# Total people who deceased
total_deceased = df['count_of_case'][df['clinical_status'] == 'Deceased (based on date of death)'].sum()
# Total people who need oxygen supplementation
total_oxygen_supplementation = df['count_of_case'][df['clinical_status'] == 'Requires Oxygen Supplementation in General Ward'].sum()
# plotting the graph
plt.figure(figsize=(20,10)) # define the size of the figure
plt.title('Percentages of reason people needing to go to the hospital', fontsize=20) # setting the title
colors = sns.color_palette('pastel')[0:5] # define the color of the pie chart
plt.pie([actual_hospitalised, total_oxygen_supplementation, total_icu, total_deceased], labels=['General Hospitalisation', 'Oxygen supplement in General Ward', 'ICU', 'Deceased'], autopct='%1.1f%%', colors=colors, wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' }) # plotting the pie chart
inner_circle = plt.Circle( (0,0), 0.7, color='white') # define the inner circle to make the pie chart look like a doughnut
p = plt.gcf() # get the current figure
p.gca().add_artist(inner_circle) # add the inner circle to the figure
plt.show() 

# Analysis of the data and graph of the data
```
