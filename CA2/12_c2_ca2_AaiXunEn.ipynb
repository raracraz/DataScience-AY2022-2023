```python
# This box is used for showing the number of hospital admissions in Singapore due to Covid in the last 2 months

# Importing libraries #
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Code starts here #
# Importing the data #
df = pd.read_csv('./csvData/new-covid-19-hospital-admissions.csv')

# Processing the data #
df['date'] = pd.to_datetime(df['date'])
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
df['new_hospital_admissions'] = df['new_hospital_admissions'].astype(int)

# Visualizing the data #
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)
plt.title('Number of daily hospital admissions in Singapore due to Covid in the last 2 months', fontsize=20)

# Plotting the data #
sns.set_style('ticks', rc={'axes.facecolor': '.9', 'axes.grid': False})
major_ticks = np.arange(0, 221, 20)
minor_ticks = np.arange(0, 210, 5)
plt.xticks(rotation=60, ha='right')
ax.set_yticks(major_ticks)
ax.set_yticks(minor_ticks, minor=True)

ax.grid(which='minor', alpha=0.5, linestyle='dashed')
ax.grid(which='major', alpha=1)

sns.despine()
sns.barplot(x='date', y='new_hospital_admissions', data=df)

plt.ylabel('Number of hospital admissions', fontsize=20)
plt.xlabel('Dates recorded', fontsize=20)

# set grid on
plt.grid(True, 'major', 'y', ls='--', lw=.5, c='k', alpha=.3)

# set a best fit line to help see the trend between dates
ind = np.arange(len(df))
z = np.polyfit(ind, df['new_hospital_admissions'], 1)
p = np.poly1d(z)
plt.plot(ind, p(ind), "r--")
plt.show()

# Code ends here

# # # # # # # # # # # # # # # # # # # # # # # # # #  Analysis of the data and graph of the data # # # # # # # # # # # # # # # # # # # # # # # # # #
# IMPORTANT EVENTS in the 2 months:                                                                                                               #
# 1. Singapore to lift most lockdown restrictions on June 19 starting from June 2, 2022                                                           #
# 2. Group size limits will go up to 10 persons and masks will no longer be mandatory in outdoor settings from March 29 onwards                   #
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
print('OBSERVATIONS about the dataset:')
# show the shape of the dataset
print('\nThis is the shape of the dataset: ' + str(df.shape))
# show the index of the dataset
print('This is the index of the dataset: ' + str(df.index))
# show the columns of the dataset
print('This is the columns of the dataset: ' + str(df.columns))
# show a summary of the dataset
print('This is a summary of the dataset: ')
print(str(df.info()))
print('\nINSIGHTS:')
print('\nThe number of hospital admissions in Singapore due to Covid in the last 2 months is: ', str(df['new_hospital_admissions'].sum()) + '.')
print('The best fit line suggests that there is currently a trend of hospital admissions increasing in Singapore in the last 2 months becausing of Covid.')
# find the average number of hospital admissions increased per day 
print('The average number of hospital admissions increased per day in the last 2 months is: ', str((df['new_hospital_admissions'].mean() / 61).round(2)) + '.')
print('A significant spike of hospital admissions can be observed from June 19, 2022 onwards, likely caused by the lifting of lockdown restrictions and the increase in group size limits with no masks mandatory in outdoor settings and the lifting of safe distance measures.')
print('In order to move forward with the impact of Covid-19 in Singapore, the government lifted most restrictions; however, this resulted in the rise in trend of hospital admissions in the last 2 months. Some researchers have debated that this return to normality is too fast, and that the covid situation will worsen again if the government does lifts the lockdown restrictions, which came true as seen in the graph above.')
print('I feel that I would reccommend the government to have delayed the lifting of restrictions to when the covid situation gets better and reaches a more stable state.')
```

```python
# Percentage of groups of people being hospitalised, Needing oxygen supplementation, Admitted into ICU and Deceased
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# Importing the data
try:
    df = pd.read_csv('./csvData/patients-needing-oxygen-supplementation-icu-care-or-died-by-age-groups.csv')
    print('Loaded dataset successfully')
except:
    print('File not found')
# Loading the data into a dataframe
df['date'] = pd.to_datetime(df['day_of_as_of_date'])
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
df['age_groups'] = df['age_groups'].astype(str)
df['clinical_status'] = df['clinical_status'].astype(str)
df['count_of_case'] = df['count_of_case'].astype(int)
# Getting the total number of people hospitalised
df = df.groupby(['clinical_status']).sum().reset_index()
# if the clinical status is 'Hospitalised', minus all other clinical statuses to find the actual number of people hospitalised
total_hospitalised = df['count_of_case'][df['clinical_status'] == 'Hospitalised'].sum()
total_not_hospitalised = df['count_of_case'][df['clinical_status'] != 'Hospitalised'].sum()
actual_hospitalised = total_hospitalised - total_not_hospitalised
# Total people admitted to ICU
total_icu = df['count_of_case'][df['clinical_status'] == 'ICU'].sum()
# Total people who deceased
total_deceased = df['count_of_case'][df['clinical_status'] == 'Deceased (based on date of death)'].sum()
# Total people who need oxygen supplementation
total_oxygen_supplementation = df['count_of_case'][df['clinical_status'] == 'Requires Oxygen Supplementation in General Ward'].sum()
# plotting the graph
plt.figure(figsize=(20,10)) # define the size of the figure
# plt.subplot(3,2,1) # define the subplot
plt.title('Percentages of patient hospital admission outcomes for all age groups', fontsize=20) # setting the title
colors = sns.color_palette('pastel')[0:5] # define the color of the pie chart
plt.pie([actual_hospitalised, total_oxygen_supplementation, total_icu, total_deceased], labels=[actual_hospitalised, total_oxygen_supplementation, total_icu, total_deceased], autopct='%1.1f%%', colors=colors, wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' }) # plotting the pie chart
inner_circle = plt.Circle( (0,0), 0.7, color='white') # define the inner circle to make the pie chart look like a doughnut
plt.legend(['General Hospitalisation', 'Oxygen supplement in General Ward', 'ICU', 'Deceased'], loc='upper left') # setting the legend
p = plt.gcf() # get the current figure
p.gca().add_artist(inner_circle) # add the inner circle to the figure
plt.axis('equal')

plt.show() 

# FOR AGE GROUP 70 YEARS OLD AND ABOVE

df2 = pd.read_csv('./csvData/patients-needing-oxygen-supplementation-icu-care-or-died-by-age-groups.csv')
# if the age_group of the patient is the same, put them in another dataframe
df2 = df2[df2['age_groups'] == '70 years old and above']
total_hospitalised = df2['count_of_case'][df2['clinical_status'] == 'Hospitalised'].sum()
total_not_hospitalised = df2['count_of_case'][df2['clinical_status'] != 'Hospitalised'].sum()
actual_hospitalised = total_hospitalised - total_not_hospitalised
total_icu = df2['count_of_case'][df2['clinical_status'] == 'ICU'].sum()
total_deceased = df2['count_of_case'][df2['clinical_status'] == 'Deceased (based on date of death)'].sum()
total_oxygen_supplementation = df2['count_of_case'][df2['clinical_status'] == 'Requires Oxygen Supplementation in General Ward'].sum()
# plotting the graph
plt.figure(figsize=(60,60.5)) # define the size of the figure
plt.subplot(6,1,1) # define the subplot
# plt.title('Percentages of reason people needing to go to the hospital', fontsize=20) # setting the title
# set the title with margin to the right
plt.title('Percentages of patient hospital admission outcomes for age groups 70 years and above', fontsize=18, pad=20)
colors = sns.color_palette('pastel')[0:5] # define the color of the pie chart
plt.pie([actual_hospitalised, total_oxygen_supplementation, total_icu, total_deceased], labels=[actual_hospitalised, total_oxygen_supplementation, total_icu, total_deceased], autopct='%1.1f%%', colors=colors, wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' }) # plotting the pie chart
inner_circle = plt.Circle( (0,0), 0.7, color='white') # define the inner circle to make the pie chart look like a doughnut
p = plt.gcf() # get the current figure
p.gca().add_artist(inner_circle) # add the inner circle to the figure
# set the legend with padding to the left
plt.legend(['General Hospitalisation', 'Oxygen supplement in General Ward', 'ICU', 'Deceased'], loc='upper left', ncol=2, borderaxespad=0)
# plt.show()

# FOR AGE GROUP 60 - 69 YEARS OLD
df3 = pd.read_csv('./csvData/patients-needing-oxygen-supplementation-icu-care-or-died-by-age-groups.csv')
# if the age_group of the patient is the same, put them in another dataframe
df3 = df3[df3['age_groups'] == '60 - 69 years old']
total_hospitalised = df3['count_of_case'][df3['clinical_status'] == 'Hospitalised'].sum()
total_not_hospitalised = df3['count_of_case'][df3['clinical_status'] != 'Hospitalised'].sum()
actual_hospitalised = total_hospitalised - total_not_hospitalised
total_icu = df3['count_of_case'][df3['clinical_status'] == 'ICU'].sum()
total_deceased = df3['count_of_case'][df3['clinical_status'] == 'Deceased (based on date of death)'].sum()
total_oxygen_supplementation = df3['count_of_case'][df3['clinical_status'] == 'Requires Oxygen Supplementation in General Ward'].sum()
# plotting the graph
# plt.figure(figsize=(20,10)) # define the size of the figure
plt.subplot(6,1,2) # define the subplot
plt.title('Percentages of patient hospital admission outcomes for age group 60 - 69 years old', fontsize=18, pad=20) # setting the title
colors = sns.color_palette('pastel')[0:5] # define the color of the pie chart
plt.pie([actual_hospitalised, total_oxygen_supplementation, total_icu, total_deceased], labels=[actual_hospitalised, total_oxygen_supplementation, total_icu, total_deceased], autopct='%1.1f%%', colors=colors, wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' }) # plotting the pie chart
inner_circle = plt.Circle( (0,0), 0.7, color='white') # define the inner circle to make the pie chart look like a doughnut
p = plt.gcf() # get the current figure
p.gca().add_artist(inner_circle) # add the inner circle to the figure
plt.legend(['General Hospitalisation', 'Oxygen supplement in General Ward', 'ICU', 'Deceased'], loc='upper left', ncol=2, borderaxespad=0)
# plt.show()

# FOR AGE GROUP 40 - 59 YEARS OLD
df4 = pd.read_csv('./csvData/patients-needing-oxygen-supplementation-icu-care-or-died-by-age-groups.csv')
# if the age_group of the patient is the same, put them in another dataframe
df4 = df4[df4['age_groups'] == '40 - 59 years old']
total_hospitalised = df4['count_of_case'][df4['clinical_status'] == 'Hospitalised'].sum()
total_not_hospitalised = df4['count_of_case'][df4['clinical_status'] != 'Hospitalised'].sum()
actual_hospitalised = total_hospitalised - total_not_hospitalised
total_icu = df4['count_of_case'][df4['clinical_status'] == 'ICU'].sum()
total_deceased = df4['count_of_case'][df4['clinical_status'] == 'Deceased (based on date of death)'].sum()
total_oxygen_supplementation = df4['count_of_case'][df4['clinical_status'] == 'Requires Oxygen Supplementation in General Ward'].sum()
# plotting the graph
# plt.figure(figsize=(20,10)) # define the size of the figure
plt.subplot(6,1,3) # define the subplot
plt.title('Percentages of patient hospital admission outcomes for age group 40 - 59 years old', fontsize=18, pad=20) # setting the title
colors = sns.color_palette('pastel')[0:5] # define the color of the pie chart
plt.pie([actual_hospitalised, total_oxygen_supplementation, total_icu, total_deceased], labels=[actual_hospitalised, total_oxygen_supplementation, total_icu, total_deceased], autopct='%1.1f%%', colors=colors, wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' }) # plotting the pie chart
inner_circle = plt.Circle( (0,0), 0.7, color='white') # define the inner circle to make the pie chart look like a doughnut
p = plt.gcf() # get the current figure
p.gca().add_artist(inner_circle) # add the inner circle to the figure
plt.legend(['General Hospitalisation', 'Oxygen supplement in General Ward', 'ICU', 'Deceased'], loc='upper left', ncol=2, borderaxespad=0)
# plt.show()

# FOR AGE GROUP 20 - 39 YEARS OLD
df5 = pd.read_csv('./csvData/patients-needing-oxygen-supplementation-icu-care-or-died-by-age-groups.csv')
# if the age_group of the patient is the same, put them in another dataframe
df5 = df5[df5['age_groups'] == '20 - 39 years old']
total_hospitalised = df5['count_of_case'][df5['clinical_status'] == 'Hospitalised'].sum()
total_not_hospitalised = df5['count_of_case'][df5['clinical_status'] != 'Hospitalised'].sum()
actual_hospitalised = total_hospitalised - total_not_hospitalised
total_icu = df5['count_of_case'][df5['clinical_status'] == 'ICU'].sum()
total_deceased = df5['count_of_case'][df5['clinical_status'] == 'Deceased (based on date of death)'].sum()
total_oxygen_supplementation = df5['count_of_case'][df5['clinical_status'] == 'Requires Oxygen Supplementation in General Ward'].sum()
# plotting the graph
# plt.figure(figsize=(20,10)) # define the size of the figure
plt.subplot(6,1,4) # define the subplot
plt.title('Percentages of patient hospital admission outcomes for age group 20 - 39 years old', fontsize=18, pad=20) # setting the title
colors = sns.color_palette('pastel')[0:5] # define the color of the pie chart
plt.pie([actual_hospitalised, total_oxygen_supplementation, total_icu], labels=[actual_hospitalised, total_oxygen_supplementation, total_icu], autopct='%1.1f%%', colors=colors, wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' }) # plotting the pie chart
inner_circle = plt.Circle( (0,0), 0.7, color='white') # define the inner circle to make the pie chart look like a doughnut
p = plt.gcf() # get the current figure
p.gca().add_artist(inner_circle) # add the inner circle to the figure
plt.legend(['General Hospitalisation', 'Oxygen supplement in General Ward', 'ICU'], loc='upper left', ncol=2, borderaxespad=0)
# plt.show()

# FOR AGE GROUP 12 - 19 YEARS OLD
df6 = pd.read_csv('./csvData/patients-needing-oxygen-supplementation-icu-care-or-died-by-age-groups.csv')
# if the age_group of the patient is the same, put them in another dataframe
df6 = df6[df6['age_groups'] == '12 - 19 years old']
total_hospitalised = df6['count_of_case'][df6['clinical_status'] == 'Hospitalised'].sum()
total_not_hospitalised = df6['count_of_case'][df6['clinical_status'] != 'Hospitalised'].sum()
actual_hospitalised = total_hospitalised - total_not_hospitalised
total_icu = df6['count_of_case'][df6['clinical_status'] == 'ICU'].sum()
total_deceased = df6['count_of_case'][df6['clinical_status'] == 'Deceased (based on date of death)'].sum()
total_oxygen_supplementation = df6['count_of_case'][df6['clinical_status'] == 'Requires Oxygen Supplementation in General Ward'].sum()
# plotting the graph
# plt.figure(figsize=(20,10)) # define the size of the figure
plt.subplot(6,1,5) # define the subplot
plt.title('Percentages of patient hospital admission outcomes for age group 12 - 19 years old', fontsize=18, pad=20) # setting the title
colors = sns.color_palette('pastel')[0:5] # define the color of the pie chart
plt.pie([actual_hospitalised, total_icu], labels=[actual_hospitalised, total_icu], autopct='%1.1f%%', colors=colors, wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' }) # plotting the pie chart
inner_circle = plt.Circle( (0,0), 0.7, color='white') # define the inner circle to make the pie chart look like a doughnut
p = plt.gcf() # get the current figure
p.gca().add_artist(inner_circle) # add the inner circle to the figure
plt.legend(['General Hospitalisation', 'Oxygen supplement in General Ward'], loc='upper left', ncol=2, borderaxespad=0)
# plt.show()

# FOR AGE GROUP 5 - 11 YEARS OLD
df7 = pd.read_csv('./csvData/patients-needing-oxygen-supplementation-icu-care-or-died-by-age-groups.csv')
# if the age_group of the patient is the same, put them in another dataframe
df7 = df7[df7['age_groups'] == '0 - 11 years old']
total_hospitalised = df7['count_of_case'][df7['clinical_status'] == 'Hospitalised'].sum()
total_not_hospitalised = df7['count_of_case'][df7['clinical_status'] != 'Hospitalised'].sum()
actual_hospitalised = total_hospitalised - total_not_hospitalised
total_icu = df7['count_of_case'][df7['clinical_status'] == 'ICU'].sum()
total_deceased = df7['count_of_case'][df7['clinical_status'] == 'Deceased (based on date of death)'].sum()
total_oxygen_supplementation = df7['count_of_case'][df7['clinical_status'] == 'Requires Oxygen Supplementation in General Ward'].sum()
# plotting the graph
# plt.figure(figsize=(20,10)) # define the size of the figure
plt.subplot(6,1,6) # define the subplot
plt.title('Percentages of patient hospital admission outcomes for age group 0 - 11 year old', fontsize=18, pad=20) # setting the title
colors = sns.color_palette('pastel')[0:5] # define the color of the pie chart
plt.pie([actual_hospitalised, total_oxygen_supplementation, total_icu], labels=[actual_hospitalised, total_oxygen_supplementation, total_icu ], autopct='%1.1f%%', colors=colors, wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' }) # plotting the pie chart
inner_circle = plt.Circle( (0,0), 0.7, color='white') # define the inner circle to make the pie chart look like a doughnut
p = plt.gcf() # get the current figure
p.gca().add_artist(inner_circle) # add the inner circle to the figure
plt.legend(['General Hospitalisation', 'Oxygen supplement in General Ward', 'ICU'], loc='upper left', ncol=2, borderaxespad=0)
plt.show()

# # # # # # # # # # # # # # # # # # # # # # # # # #  Analysis of the data and graph of the data # # # # # # # # # # # # # # # # # # # # # # # # # 1. Data in this dataset only contains data in one month from 2022-06-29 to 2022-07-26                                                       #
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # 

print('OBSERVATIONS about the dataset:')
# show the shape of the dataset
print('\nThis is the shape of the dataset: ' + str(df.shape))
# show the index of the dataset
print('This is the index of the dataset: ' + str(df.index))
# show the columns of the dataset
print('This is the columns of the dataset: ' + str(df.columns))
# show the summary
print('This is a summary of the dataset: ')
print(str(df.info()))
print('\nINSIGHTS:')
print('The first pie chart combines every age group based on their clinical status when admitted into the hospital, and shows the percentage of people who are hospitalised, ICU, and deceased. The purpose of this pie chart is to give the reader a overview of the entire situation of patients in the hospital in the past 1 months.')
print('The following pie charts shows the indivdiual situation for each age group, 70 years old and above, 60 to 69 years old, 40 to 59 years old, 20 to 39 years old, 12 to 19 years old and 0 to 11 years old respectively.')
print('This combination of pie charts clearly allows me to see which age group did Covid affect the most in resulting deaths, admitted into ICU, Oxygen supplementation, and hospitalisation. And the most vulnerable age group is the 70 years old and above age range, as the percentage of deaths is the highest at 0.6% with 66 people.')
print('With the most vulnerable age group being identified at 70 years old and above, we can also see that the deaths occured in this age group also accounted for 85.7% of all deaths recorded in the past 1 month. Thus, I recommend that special attention should be paid to the safety of the patients in this age group.')
```

```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Importing the dataset
try:
    df = pd.read_csv('./csvData/covid-hospitalizations.csv')
    print('Loaded dataset successfully!')
except:
    print('The file is not found')

# get all the countries that I want to compare with
singaporedf = df[df['entity'] == 'Singapore']
malaysiadf = df[df['entity'] == 'Malaysia']
norwaydf = df[df['entity'] == 'Norway']
belgiumdf = df[df['entity'] == 'Belgium']
isrealdf = df[df['entity'] == 'Israel']

# only keep the dates that is present in the singaporedf and indicator is present in the singapore
singaporedf = singaporedf[singaporedf['indicator'].isin(['Weekly new hospital admissions'] )]
malaysiadf = malaysiadf[malaysiadf['indicator'].isin(['Weekly new hospital admissions']) & malaysiadf['date'].isin(singaporedf['date'].unique())]
norwaydf = norwaydf[norwaydf['indicator'].isin(['Weekly new hospital admissions']) & norwaydf['date'].isin(singaporedf['date'].unique())]
belgiumdf = belgiumdf[belgiumdf['indicator'].isin(['Weekly new hospital admissions']) & belgiumdf['date'].isin(singaporedf['date'].unique())]
israeldf = isrealdf[isrealdf['indicator'].isin(['Weekly new hospital admissions']) & isrealdf['date'].isin(singaporedf['date'].unique())]

# add all the df into one df
df = pd.concat([singaporedf, malaysiadf, norwaydf, belgiumdf, israeldf], ignore_index=True)

# plot the box plot
fig = plt.figure(figsize=(20, 10)) # define the size of the figure
# plot the box plot using seaborn using singaporedf, malaysiadf, norwaydf, belgiumdf, isrealdf
boxplot = sns.boxplot(x='entity', y='value', data=df)
plt.title('Box plot of the number of hospital admissions from 2022-05-31 to 2022-07-24 to in 5 different countries', fontsize=18, pad=20) # setting the title
plt.ylabel('Number of hospital admissions', fontsize=18) # setting the y-axis label
plt.xlabel('Countries', fontsize=18) # setting the x-axis label
# set yticks from 0 - 2500 
plt.yticks(np.arange(0, 2501, 100))
# set the color palette for the box plot
sns.color_palette("crest", as_cmap=True)
# plot the striplot 
boxplot = sns.stripplot(x='entity', y='value', data=df, jitter=True, size=3, color='black')
# set the labels
plt.xlabel('Countries', fontsize=18) # setting the x-axis label
plt.ylabel('Number of hospital admissions per week', fontsize=18) # setting the y-axis label
plt.grid(True, axis='y', alpha=.8) # set the grid
plt.show()

# # # # # # # # # # # # # # # # # # # # # # # # # #  Analysis of the data and graph of the data # # # # # # # # # # # # # # # # # # # # # # # # # 1. Data in this dataset only contains data in two month from 2022-05-31 to 2022-07-24
# 2. The countries that I taken into consideration are: Singapore, Malaysia, Norway, Belgium and Israel
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # 

# Population of countries (as of 2022-7-24)
# Population of Singapore: 5,947,440
# Population of Malaysia: 33,251,353
# Population of Norway: 5,510,690
# Population of Belgium: 11,695,195
# Population of Israel: 8,946,467

print('OBSERVATIONS about the dataset:')
# show the shape of the dataset
print('\nThis is the shape of the dataset: ' + str(df.shape))
# show the index of the dataset
print('This is the index of the dataset: ' + str(df.index))
# show the columns of the dataset
print('This is the columns of the dataset: ' + str(df.columns))
# show the summary
print('This is a summary of the dataset: ')
print(str(df.info()))
print('\nINSIGHTS:')
print('The purpose of this box plot is meant to compare the number of hospital admissions from a recent time period of Singapore to 4 other different countries, namely Malaysia, Norway, Belgium and Israel.')
print('The reason why I have chosen the countries I chose is because of their location geographically. Some of the chosen countries are in different continents, thus I am able to compare the situation of Covid-19 on different continents from the date of 2022-05-31 to 2022-07-24. For example, Israel is in the Middle east, Norway and Belgium being in Europe, then Singapore and Malaysia being in South east Asia.')
```
