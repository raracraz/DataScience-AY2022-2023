```python
# This box is used for showing the number of hospital admissions in Singapore due to Covid in the last 2 months
# 
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# Code starts here
# Importing the data
df = pd.read_csv('./csvData/new-covid-19-hospital-admissions.csv')
# Processing the data
df['date'] = pd.to_datetime(df['date'])
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
df['new_hospital_admissions'] = df['new_hospital_admissions'].astype(int)
# Visualizing the data with the width of the object and the height of the object being big
plt.figure(figsize=(20,10))
sns.barplot(x='date', y='new_hospital_admissions', data=df)
# put rotation=30, ha='right' on the x axis
plt.xticks(rotation=60, ha='right')
# set grid on
plt.grid(True)
# set a best fit line
sns.lineplot(x='date', y='new_hospital_admissions', data=df)
plt.show()
# Code ends here
```
