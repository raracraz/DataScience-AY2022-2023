```python
import numpy as np
import matplotlib.pyplot as plt

# global variables: shades of blue
colors = ['#87cefa', '#1e90ff', '#4169e1', '#6495ed', '#F0E442', '#87ceeb', '#00bfff', '#ace5ee', '#ffa500', '#ff8c00', '#ff7f50', '#ff6347', '#ff4500', '#ffd700', '#ffa500', '#ff8c00']
plt.rcParams['axes.prop_cycle'] = plt.cycler('color', colors)

overall_primary_that_progress_to_secondary = [[], []]
primary_that_progress_to_secondary = np.genfromtxt("csvDataEDU/percentage-of-p1-cohort-eligible-for-secondary-school.csv", delimiter=",", skip_header=8)

overall_o_level_that_progress_to_post_secondary = [[], []]
olevel_that_progress_to_post_secondary = np.genfromtxt('csvDataEDU/percentage-of-o-level-cohort-that-progressed-to-post-secondary-education.csv', delimiter=',', skip_header=1, dtype=str)

overall_n_level_that_progress_to_post_secondary = [[], []]
nlevel_that_progress_to_post_secondary = np.genfromtxt('csvDataEDU/percentage-of-n-level-cohort-that-progressed-to-post-secondary-education.csv', delimiter=',', skip_header=1, dtype=str)

nanyang_max_cutoff_list = []
nanyang_max_cutoff = np.genfromtxt('csvDataEDU/nanyang-polytechnic-gce-o-level-aggregate-cut-off-points-by-course-2014-2021.csv', delimiter=',', skip_header=1, dtype=str)

average_of_all_students_that_progress = [[], []]
average_o_level_that_progress_to_post_secondary = [[], []]
average_n_level_that_progress_to_post_secondary = [[], []]

for i in range(len(nanyang_max_cutoff)):
    score = nanyang_max_cutoff[i][4]
    if len(score) > 2:
        score = score[-2:].strip()
        nanyang_max_cutoff_list.append(score)
    else:
        score = score.strip()
        nanyang_max_cutoff_list.append(score)
    # sort the list from lowest to highest

nanyang_max_cutoff_list.sort(key=int)

for i in range(len(primary_that_progress_to_secondary)):
    overall_primary_that_progress_to_secondary[0].append(int(primary_that_progress_to_secondary[i][0]))
    overall_primary_that_progress_to_secondary[1].append(100 - float(primary_that_progress_to_secondary[i][1]))
# remove the last element of both lists because the datasets do not match
overall_primary_that_progress_to_secondary[0].pop()
overall_primary_that_progress_to_secondary[1].pop()

for i in range(len(olevel_that_progress_to_post_secondary)):
    if olevel_that_progress_to_post_secondary[i][1] == "Overall":
        overall_o_level_that_progress_to_post_secondary[0].append(int(olevel_that_progress_to_post_secondary[i][0]))
        overall_o_level_that_progress_to_post_secondary[1].append(100 - float(olevel_that_progress_to_post_secondary[i][2]))
# remove the last element of both lists because the datasets do not match
overall_o_level_that_progress_to_post_secondary[0].pop()
overall_o_level_that_progress_to_post_secondary[1].pop()

for i in range(len(nlevel_that_progress_to_post_secondary)):
    if nlevel_that_progress_to_post_secondary[i][1] == "Overall":
        overall_n_level_that_progress_to_post_secondary[0].append(int(nlevel_that_progress_to_post_secondary[i][0]))
        overall_n_level_that_progress_to_post_secondary[1].append(100 - float(nlevel_that_progress_to_post_secondary[i][2]))
# remove the first element of both lists because the datasets do not match
overall_n_level_that_progress_to_post_secondary[0].pop(0)
overall_n_level_that_progress_to_post_secondary[1].pop(0)

# find average of all three datasets
for i in range(len(overall_primary_that_progress_to_secondary[1])):
    average_of_all_students_that_progress[0].append(int(overall_primary_that_progress_to_secondary[0][i]))
    average_of_all_students_that_progress[1].append((float(overall_primary_that_progress_to_secondary[1][i]) + float(overall_o_level_that_progress_to_post_secondary[1][i]) + float(overall_n_level_that_progress_to_post_secondary[1][i])) / 3)

print(nanyang_max_cutoff_list)
print(overall_primary_that_progress_to_secondary)
print(overall_o_level_that_progress_to_post_secondary)
print(overall_n_level_that_progress_to_post_secondary)
print(average_of_all_students_that_progress)


# plot a histogram showing the max cutoff points of Nanyang Polytechnic GCE O-level throughout the years from 2014-2021
plt.figure(figsize=(15, 10))
plt.hist(nanyang_max_cutoff_list, bins=np.arange(0, 22, 1), align='left', color='#008080', edgecolor='white', linewidth=1.5)
plt.xlabel("Cut-off points")
plt.ylabel("Frequency of Courses with that cut-off points")
plt.suptitle("Histogram", fontsize=35, fontweight="bold", y=0.934)
# shows the cut off point with the highest frequency in the histogram with a red dot, and the cut off point with the lowest frequency in the histogram with a blue dot
plt.scatter(nanyang_max_cutoff_list[np.argmax(np.bincount(nanyang_max_cutoff_list).max())], np.bincount(nanyang_max_cutoff_list).max(), color="red", marker="o")
plt.scatter(nanyang_max_cutoff_list[np.argmin(np.bincount(nanyang_max_cutoff_list).min())], np.bincount(nanyang_max_cutoff_list).min(), color="blue", marker="o")
# makes a grid 
plt.grid(True, linestyle='--', linewidth=1, color='grey', alpha=0.5)
# yticks
plt.yticks(np.arange(0, 55, 5))
plt.show()
print("This Histogram shows the trend of the cut-off points of Nanyang Polytechnic GCE O-level throughout the years from 2014-2021")
# print the cut off point with the highest frequency in the histogram, and show how many courses have that cut off point
print("The cut-off point with the highest frequency is 16 and it appears in", np.bincount(nanyang_max_cutoff_list).max(), "courses from 2014-2021")
print("The cut-off point with the lowest frequency is 6 and it appears in 1 courses from 2014-2021")
print("The highest point is marked by a red dot, and the lowest point is marked by a blue dot")
print("This data of 16 being the score with the highest frequency means that the majority of the courses no matter the year have this score, this fact that 16 is the highest frequenncy can also mean that the number of courses such as business studies, engineering, commonly averaging around the score of 16 is higher then the number of courses such as sciences and computer science, which averages around 6-13.")

# plot a line graph data
plt.figure(figsize=(15, 15))
plt.suptitle("Line graph", fontsize=35, fontweight="bold", y=0.965)
plt.subplot(2, 2, 1)
plt.plot(overall_n_level_that_progress_to_post_secondary[0], overall_n_level_that_progress_to_post_secondary[1], label="N-Level", linewidth=2, color="#E74C3C")
plt.yticks(np.arange(0, 13, 1))
plt.xticks(np.arange(overall_n_level_that_progress_to_post_secondary[0][0], overall_n_level_that_progress_to_post_secondary[0][-1]+1, 1))
plt.xlabel("Year")
plt.ylabel("Percentage of failure (%)")
plt.title("N-Level students passing 'N' levels", fontsize=15, fontweight="bold", y=1.05)
plt.legend(loc="upper right")
plt.grid(True, linestyle='--', linewidth=1, color='grey', alpha=0.5)

plt.subplot(2, 2, 2)
plt.plot(overall_primary_that_progress_to_secondary[0], overall_primary_that_progress_to_secondary[1], label="Primary", linewidth=2, color="#3498DB")
plt.yticks(np.arange(0, 13, 1))
plt.xticks(np.arange(overall_primary_that_progress_to_secondary[0][0], overall_primary_that_progress_to_secondary[0][-1]+1, 1))
plt.xlabel("Year")
plt.ylabel("Percentage of failure (%)")
plt.title("Primary students passing PSLE", fontsize=15, fontweight="bold", y=1.05)
plt.legend(loc="upper right")
plt.grid(True, linestyle='--', linewidth=1, color='grey', alpha=0.5)

plt.subplot(2, 2, 3)
plt.plot(overall_o_level_that_progress_to_post_secondary[0], overall_o_level_that_progress_to_post_secondary[1], label="O-Level", linewidth=2, color="#1D8348")
plt.yticks(np.arange(0, 13, 1))
plt.xticks(np.arange(overall_o_level_that_progress_to_post_secondary[0][0], overall_o_level_that_progress_to_post_secondary[0][-1]+1, 1))
plt.xlabel("Year")
plt.ylabel("Percentage of failure (%)")
plt.title("O-Level students passing 'O' levels", fontsize=15, fontweight="bold", y=1.05)
plt.legend(loc="upper right")
plt.grid(True, linestyle='--', linewidth=1, color='grey', alpha=0.5)

plt.subplot(2, 2, 4)
plt.plot(overall_n_level_that_progress_to_post_secondary[0], overall_n_level_that_progress_to_post_secondary[1], label="N-Level", linewidth=2, color="#E74C3C")
plt.plot(overall_primary_that_progress_to_secondary[0], overall_primary_that_progress_to_secondary[1], label="Primary", linewidth=2, color="#3498DB")
plt.plot(overall_o_level_that_progress_to_post_secondary[0], overall_o_level_that_progress_to_post_secondary[1], label="O-Level", linewidth=2, color="#1D8348")
plt.yticks(np.arange(0, 13, 1))
plt.xticks(np.arange(overall_n_level_that_progress_to_post_secondary[0][0], overall_n_level_that_progress_to_post_secondary[0][-1]+1, 1))
plt.xlabel("Year")
plt.ylabel("Percentage of failure (%)")
plt.title("All 3 levels percentage of failure in progressing", fontsize=15, fontweight="bold", y=1.05)
plt.legend(loc="upper right")
plt.grid(True, linestyle='--', linewidth=1, color='grey', alpha=0.5)
plt.show()

print("This line graph shows the percentage of failure of all 3 levels in progressing to higher education from 2008-2019")
print("This line graph provides the insight of the trend showing that the percentage of failure of all 3 levels 'Olevel', 'Primary', and 'N-Level' is decreasing from 2008-2019")
print("The percentage of failure in progressing for 'N-Level' in 2019 is", round(overall_n_level_that_progress_to_post_secondary[1][-1], 2),"%")
print("The percentage of failure in progressing for 'Primary' in 2019 is", round(overall_primary_that_progress_to_secondary[1][-1], 2),"%")
print("The percentage of failure in progressing for 'O-Level' in 2019 is", round(overall_o_level_that_progress_to_post_secondary[1][-1], 2),"%")
print("Showing that 2019 is the year with the least percentage of failure in progressing from 2008-2019")
print("This series of graphs clearly shows the trend of students performing better as the year increases from 2008-2019. On the last graph, titled 'All 3 levels failure in progressing', I compared the line graphs of all 3 data sets, it showed me that the 'N-Level' grade has the greatest improvement compared to 'O-Level' and 'Primary Levels', dropping from", round(overall_n_level_that_progress_to_post_secondary[1][0], 2),"% to", round(overall_n_level_that_progress_to_post_secondary[1][-1], 2),"%")

# plot a bar graph data of number of students that progressed to secondary school
# plt.subplot(232)
plt.figure(figsize=(15, 10))
plt.suptitle("Bar graph", fontsize=35, fontweight="bold", y=0.934)
ind = np.arange(12)
width = 0.29
plt.bar(ind + width + width, overall_n_level_that_progress_to_post_secondary[1], width, label="N-Level")
plt.bar(ind, overall_primary_that_progress_to_secondary[1], width, label="Primary")
plt.bar(ind + width, overall_o_level_that_progress_to_post_secondary[1], width, label="O-Level")
# plot the best fit line with np.polyfit
z = np.polyfit(ind, average_of_all_students_that_progress[1], 1)
p = np.poly1d(z)
plt.plot(ind, p(ind), "r--")
plt.yticks(np.arange(0, 13, 1))
plt.xticks(ind + width, overall_primary_that_progress_to_secondary[0])
plt.ylabel("Percentage (%)")
plt.xlabel("Year")
plt.legend(loc="upper right")
# plot a best fit graph of the data
plt.grid(True, linestyle='--', linewidth=1, color='grey', alpha=0.5)
plt.show()
print("This bar graph shows the percentage of failure of all 3 levels in progressing to higher education from 2008-2019")
print("This bar graph helps us to compare the difference between each of the categories easily, showing us the major changes in the data over time.")
print("This bar graph provides the insight of the trend showing that the percentage of failure of all 3 levels 'Olevel', 'Primary', and 'N-Level' is decreasing from 2008-2019")
print("The purpose of using a horizontal bar graph is to allow viewers to easily compare the 3 different groups of students and the value of each group, much easier for the viewer in terms of readibility compared to a vertical bar graph.")
print("I have plotted a best fit line to clearly show the trend of the data, giving us a clear indication that the amount of students in Singapore failing to progress their respective educations are decreasing as time goes on, suggesting that the quality of education in Singapore is getting better and better throughout the years.")

# plot a scattergraph with x-axis as the years and y-axis as the percentage
# plt.subplot(233)
plt.figure(figsize=(15, 10))
plt.suptitle("Scatter graph", fontsize=35, fontweight="bold", y=0.934)
plt.scatter(overall_primary_that_progress_to_secondary[0], overall_primary_that_progress_to_secondary[1], label="Primary", color="#E3963E")
plt.scatter(overall_o_level_that_progress_to_post_secondary[0], overall_o_level_that_progress_to_post_secondary[1], label="O-Level", color="#196F3D")
plt.scatter(overall_n_level_that_progress_to_post_secondary[0], overall_n_level_that_progress_to_post_secondary[1], label="N-Level", color="#3498DB")
plt.yticks(np.arange(0, 13, 1))
plt.xticks(np.arange(overall_o_level_that_progress_to_post_secondary[0][0], overall_o_level_that_progress_to_post_secondary[0][-1]+1, 1))
plt.xlabel("Year")
plt.ylabel("Percentage (%)")
plt.legend(loc="upper right")
plt.grid(True, linestyle='--', linewidth=1, color='grey', alpha=0.5)
plt.show()
print("This scatter graph shows the percentage of failure of all 3 levels in progressing to higher education from 2008-2019, similar to the bar graph in the previous graph")
print("However, it adds on to helping us see the relationship between the years and the percentage of failure of all 3 levels in progressing, showing how much a year is contributing to the percentage of failure of all 3 levels in progressing")
print("Giving us the insight that the percentage of failure of all 3 levels 'Olevel', 'Primary', and 'N-Level' is decreasing from 2008-2019")
print("This graph can also allow us to predict future trends in regards to the percentage of failure of all 3 levels progressing to higher education, like making the prediction that the current trend of students increasingly progressing to higher education will continue to increase in the future would not be far fetched.")

# plot a box plot with x-axis as the years and y-axis as the percentage
# plt.subplot(234)
plt.figure(figsize=(6.5, 15))
plt.suptitle("Box plot", fontsize=35, fontweight="bold", y=0.925)
plt.boxplot([overall_primary_that_progress_to_secondary[1], overall_o_level_that_progress_to_post_secondary[1], overall_n_level_that_progress_to_post_secondary[1]], labels=["Primary", "O-Level", "N-Level"], showmeans=True, meanline=True, meanprops={"color": "black"}, medianprops={"color": "red"})
plt.xticks([1, 2, 3], ["Primary", "O-Level", "N-Level"])
plt.yticks(np.arange(0, 13, 1))
plt.xlabel("Education Level")
plt.ylabel("Percentage (%)")
# show grid
plt.grid(True, linestyle='--', linewidth=1, color='grey', alpha=0.5)
plt.show()
print("This box plot shows us the minimum, maximum, mean as well as the the 1st and 3rd quartiles of the percentage of failure of all 3 levels in progressing to higher education from 2008-2019")
print("In this box plot, we can see that the smallest percentage of people in a year from 2008-2019 that failed to graduated Primary, O-Levels and N-Levels are 1.9%, 1.3% and 3.6% respectively. Likewise, we can also see the highest percentage of people in a year that failed to graduated Primary, O-Levels and N-Levels are 2.4%, 2.6% and 11.2% respectively.")
print("")


# things to improve
# > lines
```
