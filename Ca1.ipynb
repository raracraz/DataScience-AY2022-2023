```python
# shows the trend of students successfully moving on to higher education increasing each year from 2008 to 2020

import numpy as np
import matplotlib.pyplot as plt

overall_primary_that_progress_to_secondary = [[], []]
primary_that_progress_to_secondary = np.genfromtxt("csvData/percentage-of-p1-cohort-eligible-for-secondary-school.csv", delimiter=",", skip_header=8)

overall_o_level_that_progress_to_post_secondary = [[], []]
olevel_that_progress_to_post_secondary = np.genfromtxt('csvData/percentage-of-o-level-cohort-that-progressed-to-post-secondary-education.csv', delimiter=',', skip_header=1, dtype=str)

overall_n_level_that_progress_to_post_secondary = [[], []]
nlevel_that_progress_to_post_secondary = np.genfromtxt('csvData/percentage-of-n-level-cohort-that-progressed-to-post-secondary-education.csv', delimiter=',', skip_header=1, dtype=str)

for i in range(len(primary_that_progress_to_secondary)):
    overall_primary_that_progress_to_secondary[0].append(int(primary_that_progress_to_secondary[i][0]))
    overall_primary_that_progress_to_secondary[1].append(100 - float(primary_that_progress_to_secondary[i][1]))
# remove the last element of both lists because the datasets do not match
overall_primary_that_progress_to_secondary[0].pop()
overall_primary_that_progress_to_secondary[1].pop()

for i in range(len(olevel_that_progress_to_post_secondary)):
    if olevel_that_progress_to_post_secondary[i][1] == "Overall":
        overall_o_level_that_progress_to_post_secondary[0].append(int(olevel_that_progress_to_post_secondary[i][0]))
        overall_o_level_that_progress_to_post_secondary[1].append(100 - float(olevel_that_progress_to_post_secondary[i][2]))
# remove the last element of both lists because the datasets do not match
overall_o_level_that_progress_to_post_secondary[0].pop()
overall_o_level_that_progress_to_post_secondary[1].pop()

for i in range(len(nlevel_that_progress_to_post_secondary)):
    if nlevel_that_progress_to_post_secondary[i][1] == "Overall":
        overall_n_level_that_progress_to_post_secondary[0].append(int(nlevel_that_progress_to_post_secondary[i][0]))
        overall_n_level_that_progress_to_post_secondary[1].append(100 - float(nlevel_that_progress_to_post_secondary[i][2]))
# remove the first element of both lists because the datasets do not match
overall_n_level_that_progress_to_post_secondary[0].pop(0)
overall_n_level_that_progress_to_post_secondary[1].pop(0)

print(overall_primary_that_progress_to_secondary)
print(overall_o_level_that_progress_to_post_secondary)
print(overall_n_level_that_progress_to_post_secondary)

# plot a histogram showing n-level students that progressed to post-secondary


# plot a line graph data

plt.figure(figsize=(15, 15))
plt.subplot(2, 2, 1)
plt.plot(overall_n_level_that_progress_to_post_secondary[0], overall_n_level_that_progress_to_post_secondary[1], label="N-Level", linewidth=2, color="red")
plt.yticks(np.arange(0, 13, 1))
plt.xticks(np.arange(overall_n_level_that_progress_to_post_secondary[0][0], overall_n_level_that_progress_to_post_secondary[0][-1]+1, 1))
plt.xlabel("Year")
plt.ylabel("Percentage (%)")
plt.legend(loc="upper right")

plt.subplot(2, 2, 2)
plt.plot(overall_primary_that_progress_to_secondary[0], overall_primary_that_progress_to_secondary[1], label="Primary", linewidth=2, color="blue")
plt.yticks(np.arange(0, 13, 1))
plt.xticks(np.arange(overall_primary_that_progress_to_secondary[0][0], overall_primary_that_progress_to_secondary[0][-1]+1, 1))
plt.xlabel("Year")
plt.ylabel("Percentage (%)")
plt.legend(loc="upper right")

plt.subplot(2, 2, 3)
plt.plot(overall_o_level_that_progress_to_post_secondary[0], overall_o_level_that_progress_to_post_secondary[1], label="O-Level", linewidth=2, color="green")
plt.yticks(np.arange(0, 13, 1))
plt.xticks(np.arange(overall_o_level_that_progress_to_post_secondary[0][0], overall_o_level_that_progress_to_post_secondary[0][-1]+1, 1))
plt.xlabel("Year")
plt.ylabel("Percentage (%)")
plt.legend(loc="upper right")

plt.subplot(2, 2, 4)
plt.plot(overall_n_level_that_progress_to_post_secondary[0], overall_n_level_that_progress_to_post_secondary[1], label="N-Level", linewidth=2, color="red")
plt.plot(overall_primary_that_progress_to_secondary[0], overall_primary_that_progress_to_secondary[1], label="Primary", linewidth=2, color="blue")
plt.plot(overall_o_level_that_progress_to_post_secondary[0], overall_o_level_that_progress_to_post_secondary[1], label="O-Level", linewidth=2, color="green")
plt.yticks(np.arange(0, 13, 1))
plt.xticks(np.arange(overall_n_level_that_progress_to_post_secondary[0][0], overall_n_level_that_progress_to_post_secondary[0][-1]+1, 1))
plt.xlabel("Year")
plt.ylabel("Percentage (%)")
plt.legend(loc="upper right")

# plot a bar graph data of number of students that progressed to secondary school
# plt.subplot(232)
plt.figure(figsize=(15, 10))
ind = np.arange(12)
width = 0.3
p3 = plt.bar(ind + width + width, overall_n_level_that_progress_to_post_secondary[1], width, label="N-Level")
p1 = plt.bar(ind, overall_primary_that_progress_to_secondary[1], width, label="Primary")
p2 = plt.bar(ind + width, overall_o_level_that_progress_to_post_secondary[1], width, label="O-Level")
plt.yticks(np.arange(0, 13, 1))
plt.xticks(ind + width, overall_primary_that_progress_to_secondary[0])
plt.ylabel("Percentage (%)")
plt.xlabel("Year")
plt.legend(loc="upper right")
plt.show()

# plot a scattergraph with x-axis as the years and y-axis as the percentage
# plt.subplot(233)
plt.figure(figsize=(15, 10))
plt.scatter(overall_primary_that_progress_to_secondary[0], overall_primary_that_progress_to_secondary[1], label="Primary")
plt.scatter(overall_o_level_that_progress_to_post_secondary[0], overall_o_level_that_progress_to_post_secondary[1], label="O-Level")
plt.scatter(overall_n_level_that_progress_to_post_secondary[0], overall_n_level_that_progress_to_post_secondary[1], label="N-Level")
plt.yticks(np.arange(0, 13, 1))
plt.xticks(np.arange(overall_o_level_that_progress_to_post_secondary[0][0], overall_o_level_that_progress_to_post_secondary[0][-1]+1, 1))
plt.xlabel("Year")
plt.ylabel("Percentage (%)")
plt.legend(loc="upper right")
plt.show()

# plot a box plot with x-axis as the years and y-axis as the percentage
# plt.subplot(234)
plt.figure(figsize=(6.5, 15))
plt.boxplot([overall_primary_that_progress_to_secondary[1], overall_o_level_that_progress_to_post_secondary[1], overall_n_level_that_progress_to_post_secondary[1]])
plt.xticks([1, 2, 3], ["Primary", "O-Level", "N-Level"])
plt.yticks(np.arange(0, 13, 1))
plt.xlabel("Education Level")
plt.ylabel("Percentage (%)")
plt.show()

# plot a histogram with x-axis as primary and y-axis as percentage
# plt.subplot(235)
plt.figure(figsize=(15, 10))
plt.hist(overall_n_level_that_progress_to_post_secondary[1], bins=np.arange(0, 13, 1), label="Primary ranges")
plt.xlabel("Percentages (%)")
plt.ylabel("Values")
plt.legend(loc="upper right")
plt.show()


# things to improve
# > lines
# > change float to int for Years
```
