```python
# shows the trend of p1 cohort eligible for secondary schools and p1 cohort that did not complete secondary schools are decreasing as time goes by

import numpy as np
import matplotlib.pyplot as plt

p1_eligible_for_secondary = np.genfromtxt("csvData/percentage-of-p1-cohort-eligible-for-secondary-school.csv", delimiter=",", skip_header=3)

for i in range(len(p1_eligible_for_secondary)):
    p1_eligible_for_secondary[i][1] = 100 - p1_eligible_for_secondary[i][1]

p1_did_not_complete_secondary = np.genfromtxt('csvData/percentage-of-p1-cohort-that-did-not-complete-secondary-education.csv', delimiter=',', skip_header=1)

plt.title('p1 cohort that failed to moved on to secondary education and post-secondary education are decreasing over time.')
plt.ylabel('Percentage')
plt.xlabel('Year')

plt.plot(p1_did_not_complete_secondary[:,0], p1_did_not_complete_secondary[:,1])
plt.plot(p1_eligible_for_secondary[:,0], p1_eligible_for_secondary[:,1])

# things to improve
# > lines
# > change float to int for Years
```
